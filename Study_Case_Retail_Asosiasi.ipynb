```python
# Mengubah InvoiceDate ke Datetime
df_jateng['InvoiceDate']=pd.to_datetime(df_jateng['InvoiceDate'])
```

```python
df_jateng.head()
```

Memulai proses Transformasi data dari data yg berbentuk transaksional ke bentuk Flat table
- Tahap 1: mengubah hal yg menjadi prioritas pada keranjang belanja

```python
# Clean up spaces in product description, product category and remove any rows that don't have a valid invoice
df_jateng['PRODUCT']=df_jateng['PRODUCT'].str.strip()
df_jateng['PRODUCT_CATEGORY']=df_jateng['PRODUCT_CATEGORY'].str.strip()
```

```python
# menghapus variabel inoviceNO yang diawali dengan huruf C pada invoice numbernya
df_jateng.dropna(
    axis=0,
    subset=['InvoiceNo'],
    inplace=True
)
df_jateng['InvoiceNo']=df_jateng['InvoiceNo'].astype('str')
df_jateng=df_jateng[~(df_jateng['InvoiceNo'].str[0]=='C')]
```

menghapus variabel invoiceNo yang kosong, dan menghilangkan Transaksi yang merupakan transaksi kredit (Di awali dengan hufuf C pada invoice number)

```python
df_jateng
```

```python
# Keranjang Belanja
basket_jateng=(df_jateng[df_jateng['PROVINSI']=='Jawa Tengah']\
                .groupby(['InvoiceNo','PRODUCT_CATEGORY'])['Quantity']\
                .count()\
                .unstack().reset_index().fillna(0)\
                .set_index('InvoiceNo'))
basket_jateng.head()
```

```python
# Show a subset of columns
basket_jateng.iloc[:,[0,1,2,3,4,5,6,7]].head()
```

```python
# Melakukan proses Encoded
# Mengubah data kebentuk angka, agar sistem atau komputer dapat memahami informasi dari dataset
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_jateng_set=basket_jateng.applymap(encode_units)
basket_jateng_set.head()
```

melakukan encoding, dimana jika barang kurang dari sama dengan 0 maka keranjang tersebut bernilai 0 dan jika lebih dari 1 maka nilainya adalah 1, sehingga jika sebuah nota membeli barang A sebanyak 10 buah maka hanya akan dihitung 1. Karena analisis yang di gunakan menyaratkan seperti itu.

### Data Modelling with Association

```python
!pip install apyori
```

```python
from apyori import apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import datetime
```

```python
frequent_itemset=apriori(
    basket_jateng_set,
    min_support=0.1,
    use_colnames=True
)
frequent_itemset
```

Dimana perintah yang di gunakan adalah apriori, dengan data dari ***basket_jateng_set*** dengan minimum nilai support 0.1/ 10%.

```python
rule_jateng=association_rules(
    frequent_itemset,
    metric='lift',
    min_threshold=1
)
rule_jateng.head()
```

```python
result_jateng = rule_jateng[(rule_jateng['lift'] >= 1) &
               (rule_jateng['confidence'] >= 0.8)]

apr_result_jateng = result_jateng.sort_values(
    by='confidence',
    ascending=False
)
apr_result_jateng.head()
```

Melakukan filter untuk nilai lift ratio lebih dari sama dengan 1 dengan tingkat confidence minimal 0.8 (lebih dari sama dengan 80%).

Dengan menggunakan ambang batas ini, program memilih aturan asosiasi yang memiliki korelasi positif yang kuat (lift >= 1) dan yang memiliki tingkat kepercayaan yang tinggi (confidence >= 0.8). Ini membantu dalam mengidentifikasi aturan yang paling relevan dan kuat dalam dataset, yang dapat digunakan untuk pengambilan keputusan dan rekomendasi.

## Challenge
Tampilkan visualisasi diagram asosiasi yang paling representatif

Berdasarkan data yang diberikan, visualisasi yang paling cocok untuk aturan asosiasi dari produk-produk yang dibeli secara bersamaan adalah visualisasi heatmap. Visualisasi heatmap adalah jenis visualisasi yang menampilkan data dalam bentuk matriks, dengan setiap sel dalam matriks mewakili nilai data. Nilai data dapat diwakili dengan warna atau ukuran.

Dalam kasus ini, visualisasi heatmap dapat digunakan untuk menampilkan hubungan antara produk-produk yang dibeli secara bersamaan. Setiap sel dalam matriks akan mewakili sebuah aturan asosiasi, dengan nilai confidence sebagai nilai data. Nilai confidence yang tinggi akan ditampilkan dengan warna yang lebih gelap, sedangkan nilai confidence yang rendah akan ditampilkan dengan warna yang lebih terang.

```python
# Sort the results based on confidence in descending order
apr_result_jateng = apr_result_jateng.sort_values(
    by='confidence',
    ascending=False
)
heatmap_data = apr_result_jateng.pivot(
    index='antecedents',
    columns='consequents',
    values='confidence'
)
plt.figure(figsize=(12, 15))
sns.heatmap(
    heatmap_data,
    cmap='Greens',
    annot=True,
    fmt='.2f',
    cbar=False
)
plt.title('Association Rules Heatmap')
plt.xlabel('Consequents')
plt.ylabel('Antecedents')
plt.show()
```

Program di atas digunakan untuk membuat visualisasi heatmap yang menunjukkan association rules berdasarkan confidence. Langkah pertama adalah mengurutkan hasil dari algoritma Apriori berdasarkan confidence dalam urutan menurun. Kemudian, aturan asosiasi diambil dari hasil tersebut dan diubah menjadi tipe data string. Data aturan asosiasi yang sudah dibersihkan digunakan untuk membuat pivot table yang akan digunakan sebagai basis pembuatan heatmap. Heatmap ini mewakili association rules dalam bentuk warna hijau, di mana semakin tinggi confidence, semakin gelap warna hijau pada heatmap.
